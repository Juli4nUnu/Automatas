```python
###########################
# NO MODIFICAR ESTA CELDA #
###########################

def grade_1():
    d = punto1()
    valid_words = ['ca1xcb',
'cca10yccb',
'ca9xcb',
'ccca8zcccb',
'ca3zcb','ccca11ycccb', 'ccca12ycccb']


    invalid_words = ['caa3yb',
'ca22zb',
'azb',
'czzb',
'a8z',
]

    for i in valid_words:
        if d.accepts_input(i) != True:
            return False
    for j in invalid_words:
        if d.accepts_input(j) != False:
            return False
    return True

if grade_1():
  print('MUY BIEN! El diseño es correcto')
else:
  print('MAL! El diseño es incorrecto. Sigue intentando')
```
